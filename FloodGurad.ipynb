```python
import os
import shutil
from tqdm import tqdm

# Function to move files from source to target
def move_files(source_dir, target_dir):
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    files = os.listdir(source_dir)
    for file in tqdm(files, desc=f"Moving files from {source_dir} to {target_dir}"):
        shutil.move(os.path.join(source_dir, file), os.path.join(target_dir, file))


```

```python
import math
import os
import shutil
from tqdm import tqdm

# Updated function to move files from a starting point with a specified percentage
def move_files_from_start(source_dir, target_dir, start_index, percentage):
    # print('starting from:',start_index)
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)
    files = sorted(os.listdir(source_dir))
    total_files = len(files)
    if start_index >= total_files:
        return "Start index exceeds the number of files."

    num_files_to_move = math.ceil(total_files * percentage / 100)
    end_index = min(start_index + num_files_to_move, total_files)

    for file in tqdm(files[start_index:end_index], desc=f"Moving files from {source_dir} to {target_dir}"):
        shutil.move(os.path.join(source_dir, file), os.path.join(target_dir, file))

    return f"Moved files from index {start_index} to {end_index - 1}."
```

```python
import cv2
import json
import numpy as np
from PIL import Image, ImageDraw
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import os

# Train YOLOv8 on the custom-labeled dataset for segmentation
dataset_yaml_path = '/content/obj3.yaml'  # Define path to your dataset configuration

# تحديد المسارات
image_path = '/content/Flood3/Flood/valid/images/DJI_20231228200428_0035_V_JPG.rf.b12a164f44c0d6b290d49b3a1a6c9afb.jpg'
annotations_path = '/content/Flood3/Flood/annotations/valid_annotations.json'


# تحميل الصورة ومعرفة حجمها باستخدام OpenCV
image = cv2.imread(image_path)
height, width, channels = image.shape
print(f"حجم الصورة (العرض, الارتفاع): {width}x{height}")
# قراءة ملف JSON
with open(annotations_path, 'r') as file:
    annotations = json.load(file)

def create_mask_from_annotations(annotations, image_shape):
    mask_img = Image.new('L', (image_shape[1], image_shape[0]), 0)
    for annotation in annotations['annotations']:
        if 'segmentation' in annotation:
            polygons = annotation['segmentation']
            ImageDraw.Draw(mask_img).polygon(polygons[0], outline=1, fill=1)
    return np.array(mask_img)

# إنشاء القناع الحقيقي
ground_truth_mask = create_mask_from_annotations(annotations, (height, width))

# استخراج القناع المتوقَّع من النموذج
segmentation_results = yolo_model_segment.predict(source=image_path)
predicted_mask = segmentation_results[0].masks.data[0].cpu().numpy()
predicted_mask = cv2.resize(predicted_mask, (width, height), interpolation=cv2.INTER_NEAREST)
predicted_mask = (predicted_mask > 0.5).astype(np.uint8)  # تحويل القناع إلى ثنائي



# حساب المقاييس المطلوبة
def compute_segmentation_metrics_avg(predicted, ground_truth, num_classes):
    total_pixels = np.prod(ground_truth.shape)

    # Pixel Accuracy
    pixel_accuracy = np.sum(predicted == ground_truth) / total_pixels

    iou_list = []
    precision_list = []
    recall_list = []
    f1_list = []
    weight_list = []

    ground_truth_flat = ground_truth.flatten()
    predicted_flat = predicted.flatten()

    for class_id in range(num_classes):
        TP = np.sum((predicted_flat == class_id) & (ground_truth_flat == class_id))
        FP = np.sum((predicted_flat == class_id) & (ground_truth_flat != class_id))
        FN = np.sum((predicted_flat != class_id) & (ground_truth_flat == class_id))

        iou = TP / (TP + FP + FN) if (TP + FP + FN) > 0 else 0
        precision = TP / (TP + FP) if (TP + FP) > 0 else 0
        recall = TP / (TP + FN) if (TP + FN) > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        iou_list.append(iou)
        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)

        # for each class
        total_pixels_per_class = np.sum(ground_truth_flat == class_id)
        weight_list.append(total_pixels_per_class / total_pixels)


    mean_iou = np.sum(np.array(iou_list) * np.array(weight_list))
    mean_precision = np.sum(np.array(precision_list) * np.array(weight_list))
    mean_recall = np.sum(np.array(recall_list) * np.array(weight_list))
    mean_f1 = np.sum(np.array(f1_list) * np.array(weight_list))

    print(f"Pixel Accuracy: {pixel_accuracy:.4f}")
    print(f"Mean IoU: {mean_iou:.4f}")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-Score: {mean_f1:.4f}")

compute_segmentation_metrics_avg(predicted_mask, ground_truth_mask, num_classes=2)

# عرض القناع الحقيقي والمتوقَّع
plt.subplot(1, 2, 1)
plt.imshow(ground_truth_mask, cmap='gray')
plt.title("Ground Truth Mask")
plt.subplot(1, 2, 2)
plt.imshow(predicted_mask, cmap='gray')
plt.title("Predicted Mask")
plt.show()
```
